3. Compare and contrast mutable and immutable objects in Python with examples

->  In Python, mutable objects can be modified after creation, while immutable objects cannot. Mutable objects, like lists and dictionaries, allow for in-place changes without creating new objects. Immutable objects, like strings and tuples, require new objects to be created when modifications are made.

Mutable Objects:
Definition: Objects that can be altered after they've been created.
Examples: Lists, dictionaries, and sets.
Memory: Generally more efficient in memory as they don't require creating new objects for changes.
Modification: Modifying a mutable object doesn't change its memory address; the changes are made in place.

Immutable Objects:
Definition: Objects whose value cannot be changed once they are created.
Examples: Integers, floats, strings, tuples, and frozensets.
Memory: May use more memory as new objects are created when modifications are needed.
Modification: Modifying an immutable object creates a new object with the modified value, and any references to the original object remain unchanged.

-> In Python, type casting, also known as type conversion, is the process of converting a variable from one data type to another. This can be done implicitly (automatically by Python) or explicitly (by the programmer). Explicit type casting uses functions like int(), float(), and str() to change a variable's data type.


Explicit Type Casting Examples:
integer to float.
Python

    num_int = 5
    num_float = float(num_int)

String to Integer.
Python

    num_str = "10"

    print(num_int)  # Output: 10
Float to Integer.
Python

    num_float = 3.14
    num_int = int(num_float)

Integer to String.
Python

    num_int = 7

6.How do conditional statements work in Python? Illustrate with examples

-> Conditional statements in Python allow you to control the flow of your program's execution based on whether a condition is true or false. The most common conditional statement is the if statement, which executes a block of code only if a specified condition is met. You can also use else and elif (else if) statements to handle different scenarios when the initial condition is false.
i). if statement:
Python

age = 20

if age >= 18:
  print("You are eligible to vote.")
Explanation:
The if statement checks if the condition age >= 18 is true.
If the condition is true, the code block indented under the if statement is executed, printing "You are eligible to vote."
If the condition is false, the code block is skipped.
ii). if-else statement:
Python

temperature = 25

if temperature > 30:
  print("It's hot outside.")
else:
  print("It's not too hot.")
Explanation:
The if statement checks if the condition temperature > 30 is true.
If it's true, the first print statement is executed.
If it's false (the temperature is not greater than 30), the else statement's code block is executed, printing "It's not too hot."
iii). if-elif-else statement:
Python

score = 85

if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
else:
  print("D or F")
Explanation:
The if statement checks if the condition score >= 90 is true.
If true, "A" is printed.
If false, the elif statement checks score >= 80.
If true, "B" is printed.
If false, the next elif statement checks score >= 70, and so on.
If none of the if or elif conditions are true, the else statement's code block is executed, printing "D or F".
iv). Conditional Expressions (Ternary Operator):
Python

result = "Odd" if number % 2 != 0 else "Even"
print(result)
Explanation:
This is a concise way to write a simple if-else statement in one line.
The expression number % 2 != 0 is evaluated.
If true, the value before the else ( "Odd") is returned.
If false, the value after the else ( "Even") is returned.

7.  Describe the different types of loops in Python and their use cases with examples.

-> In Python, there are two main types of loops: for loops and while loops. for loops are used when you know the number of iterations beforehand, or when iterating over a sequence like a list or string. while loops, on the other hand, are used when you want to repeat a block of code as long as a condition is true. Nested loops are loops within loops, enabling complex iteration patterns.
i). For Loop:
Use Case: Iterating over a sequence, such as a list, tuple, or string.
Example:
Python

    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print(fruit)
This code iterates through the fruits list and prints each fruit.
ii). While Loop:
Use Case: Executing a block of code repeatedly as long as a condition is true.
Example:
Python

    count = 0
    while count < 5:
        print(count)
        count += 1
This loop prints the value of count until it reaches 5.
iii). Nested Loops:
Use Case: Creating complex iteration patterns, often involving multiple levels of iteration.
Example:
Python

    for i in range(3):
        for j in range(2):
            print(f"i: {i}, j: {j}")
This code prints all combinations of i from 0 to 2 and j from 0 to 1.
Control Statements in Loops:
break: Exits the loop immediately.
continue: Skips the current iteration and goes to the next one.
pass: A null statement, used when a statement is syntactically required but no action is needed.
