1.  **Automatic Memory Management (Garbage Collection)**

    -   **The garbage collector identifies and removes objects that are

    -   **When a variable or object is created, memory is allocated for

    -   **Each object in Python has a reference count that tracks the

    -   **When the reference count drops to zero (i.e., no more
        references), the object is automatically deallocated.**

    -   **Python uses cyclic garbage collection to handle circular
        references (objects referencing each other).**

The gc module provides developers the ability to interact with the
